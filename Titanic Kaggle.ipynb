```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
data['Age'] = data[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
data.isnull().sum()
```

```python
data.loc[data['Fare'].isnull(),'Fare']=np.mean(data['Fare'])
```

```python
tt_train=data[data['data']=='train']
del tt_train['data']
tt_test=data[data['data']=='test']
del tt_test['data']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(tt_train.drop(['Survived','PassengerId'],axis=1), 
                                                    tt_train['Survived'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,predictions)
accuracy
```
